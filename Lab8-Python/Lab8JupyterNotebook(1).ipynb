```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading a dataset:
iris_dataframe = pd.read_csv('IrisDataset/IrisDataset/Iris.csv')
```

1. Line Plot


```python
plt.plot(iris_dataframe["Id"], iris_dataframe['SepalWidthCm'])
plt.title("Iris Sepal Length")
plt.xlabel("Id")
plt.ylabel("Sepal Width (cm)")
```

```python
# adding gird-lines
plt.grid(True)

plt.plot(iris_dataframe["Id"], iris_dataframe['SepalLengthCm'])
plt.title("Iris Sepal Length")
plt.xlabel("Id")
plt.ylabel("Sepal Length (cm)")
# Set x-axis for every 10 units
plt.xticks(range(0, max(iris_dataframe["Id"]) + 1, 20))  
plt.show()
```

2. Scatter Plot


```python
plt.grid(True)
plt.scatter(iris_dataframe["SepalLengthCm"], iris_dataframe["SepalWidthCm"])
plt.title("Iris Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
```

3. Multivariate


```python
# Species
# Iris-setosa        50
# Iris-versicolor    50
# Iris-virginica     50

iris_species_color = {"Iris-setosa": "yellow", 
                      "Iris-versicolor": "red", 
                      "Iris-virginica":"blue"}

for species, species_color in iris_species_color.items():
    # creating a data frame for a specific value for 'Species' column:
    species_dataframe = iris_dataframe[iris_dataframe["Species"] == species]
    # parameter c - color
    plt.scatter(species_dataframe["PetalLengthCm"], species_dataframe["PetalWidthCm"], c = species_color, label=species)

plt.grid(True)
plt.title("Iris Sepal Length vs Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.legend()
```

4. Histogram
   

```python
sns.histplot(data=iris_dataframe, x="SepalLengthCm")
plt.title("Sepal Length Distribution")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
```

```python
sns.histplot(data=iris_dataframe, x="SepalLengthCm")
plt.title("Sepal Length Distribution")
plt.grid(axis="y")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Count")
plt.yticks(range(0, 30, 2))
```
